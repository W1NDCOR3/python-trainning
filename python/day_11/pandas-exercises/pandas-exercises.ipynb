```python
import pandas as pd
import seaborn as sns
from scipy.stats import norm 
import matplotlib.pyplot as plt
import numpy as np


csv_file_path = r"C:\Users\FORMACAO\Documents\this_is_the_repo\python-trainning\python\day_10\examples\simple_pandas\inputs\employees.csv"


# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)


df['Department'] =df['Department'].apply(lambda x: 'BPCE-IT' if not x else x)

a = np.arange(-3,3,0.1)


print(a)

normalizing = norm.pdf(a)
normalizing_2 = norm.pdf(a, 1,0.5)

print(normalizing)




axes = plt.axes()
axes.set_xlim(left = -5, right = 5)
axes.set_ylim(bottom = 0,top = 1.0)
axes.set_xticks(np.arange(-5,6,2))
axes.set_yticks(np.arange(0,1.1,0.1))
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
axes.grid()
axes.grid(linestyle = "dashed", linewidth=0.5)

plt.plot(a,normalizing)
plt.plot(a,normalizing_2, "r:")


plt.show

```
