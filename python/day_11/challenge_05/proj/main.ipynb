```python
import pandas as pd

import seaborn as sns

csv_file_path = r"C:\Users\FORMACAO\Documents\this_is_the_repo\python-trainning\python\day_11\challenge_05\proj\input\offers.csv"


# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

print(df.head(10))

def filter_data(df: pd.DataFrame) -> pd.Series:
   
    df_slice = df[df["offer_type"] == "bogo"]
    new_df = (
        df_slice.copy()
    ) 
    return new_df


def calculate_total_reward(df):
    df['total_reward'] = df['reward'] * df['difficulty']
    return df

def calculate_daily_reward(df):
    df['reward_per_day'] = df['total_reward'] // df['duration']
    return df

filtered_df = filter_data(df)

filtered_df.head()



ndf = calculate_total_reward(filtered_df)

print(ndf)

ndf = calculate_daily_reward(ndf)

print(ndf)

ndf.to_csv('new_output.csv', index=True)



```

```python

sns.set_theme(style='darkgrid') 
sns.displot(ndf['reward_per_day'])
sns.displot(ndf['total_reward'])

sns.jointplot(x='reward_per_day', y='total_reward', data=ndf)
```
